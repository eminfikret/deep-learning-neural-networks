```python
import os
import pandas as pd

from skimage.io import imread
from joblib import dump
```

```python
DATASET_DIR = "../dataset"
ARTIFACTS_DIR = "../artifacts"
OUTPUT_FILE = os.path.join(ARTIFACTS_DIR, "images_df.joblib")
```

```python
rows = []
for fname in os.listdir(DATASET_DIR):
    if not fname.lower().endswith(".png"):
        continue
    label = fname.split("_")[0]
    path = os.path.join(DATASET_DIR, fname)
    image = imread(path)
    rows.append({
        "Image": image,
        "File": fname,
        "Label": label
    })

images_df = pd.DataFrame(rows)
print(f"Loaded {len(images_df)} images with {images_df['Label'].nunique()} labels")
```

```python
images_df.head()
```

```python
dump(images_df, OUTPUT_FILE)
print(f"Saved DataFrame to {OUTPUT_FILE}")
```
